```python
import json
import os
import numpy as np
from PIL import Image
from pathlib import Path

datapath = '/SSD_DISK/users/yangzeyu/gaussian-splatting-4d/data/flame_salmon_llff_dynamic'
extension = '.png'
for f_name in ['transforms_train.json', 'transforms_test.json', 'transforms_val.json']:
    file = os.path.join(datapath, f_name)
    if os.path.exists(file):
        with open(file) as json_file:
            contents = json.load(json_file)
        frames = contents["frames"]
        for frame in frames:
            cam_name = os.path.join(datapath, frame["file_path"] + extension)
            image_path = os.path.join(datapath, cam_name) # .replace('hdImgs_unditorted', 'hdImgs_unditorted_rgba').replace('.jpg', '.png')
            image_name = Path(cam_name).stem
            image = Image.open(image_path)
            height, width = image.size[1], image.size[0]
            frame['height'] = height
            frame['width'] = width
        with open(file, 'w') as out_file:
            json.dump(contents, out_file, indent=4)
```
